# As we find that pest max is 322 & pest min is 0 & we have to divide it into 4 categories like 0-80 as low, 80-160 as medium, 160-240 as high & 240-330 as very high.

```python
label=['low','medium','high','very high']
category=[0.,80.,160.,240.,330.]
final['Pest_label']=pd.cut(final['Pest Count'],labels=label,bins=category)
```

# Modelling
1) Input parameters = Location, RH avg, Tavg

2) Change the location into numerical classes

```python
final['Location'].replace('Faridkot',0,inplace=True)
final['Location'].replace('Sirsa',1,inplace=True)
final['Location'].replace('Hisar',2,inplace=True)
```

```python
model=final[['Location','Tavg','RH avg','Pest_label']]
```

```python
model.isnull().sum()
```

```python
model.dropna(inplace=True)
```

```python
X=model.iloc[:,:-1].values
```

```python
y=model['Pest_label']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=17)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
